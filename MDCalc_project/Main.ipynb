```python
#MDCalc.verify function uses the information below to verify sepsis

information="""
patient vitals:
Temperature =39°C
Heart rate =120 bpm
Respiratory rate = 20 bpm
 WBC count =15 K/microliter
 
Doctor Diagnosis = Sepsis

"""
MDCalc.verify(information=information)
```

```python
#MDCalc.fill_form function only fills a form given the information about the patient
#requires option_number of the option found on first page of MDCalc.
#option number can be found by running the code in the cell above

#Here we are filling information in the form which 
# corresponds to the option 'Option#692:Tokyo Guidelines for Acute Cholecystitis 2018(Acute cholecystitis diagnosis and grading.)'

information1="""patient vitals:
patient exhibits murphy's sign and tenderness
has fever,elevated crp and elevated wbc count
imaging findings are not charateristic of acute cholecystitis

"""
MDCalc.form_fill(information=information1,option_number=692)#requires option number which can be found from the cell above
```
